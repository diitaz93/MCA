#Ciclo solar

```python
data=genfromtxt("monthrg.dat")
```

```python
year=data[:,0] + (data[:,1]-1.0)/12.0
solar=data[:,3]
figure()
plot(year,solar,"r+")
show()
```

```python
figure()
newx=year[solar>0]
newy=solar[solar>0]
plot(newx,newy)
show()
```

```python
from scipy import optimize as op
```

```python
def func (a,b,c,d,th):
    return a*sin(b*th+c)+d
p1,p2=op.curve_fit(func,newx,newy)
```

```python
print p1
```

```python
figure()
plot(newx,func(p1[0],p1[1],p1[2],p1[3],newx))
show()
```
